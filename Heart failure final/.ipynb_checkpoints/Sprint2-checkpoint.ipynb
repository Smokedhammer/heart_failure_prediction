## Train Test Split

Let's start by splitting our data into a training set and test set 

```python
from sklearn.model_selection import train_test_split #to split the data in training and testing part.
```

```python
features = df.drop(['DEATH_EVENT'],axis = 1)
```

```python
labels = df['DEATH_EVENT']
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=2) #variation in selecting train test records. (random state)
```

## Scaling The Data

```python
from sklearn.preprocessing import MinMaxScaler 
```

```python
scaler = MinMaxScaler((0,1))
scaler.fit(X_train)
```

```python
arr = scaler.transform(X_train)
```

```python
arr.shape #239 = number of records used for training and 12 are the number of columns. (after dropping death event.)
```

```python
X_trained_scaled = pd.DataFrame(arr,columns=X_train.columns)
X_trained_scaled
```

```python
X_train.columns
```

```python
scaler.fit(X_test)
X_test_scaled = scaler.transform(X_test)
```

```python
X_test_scaled.shape #60 = number of records used for testing and 12 are the number of columns. (after dropping death event.)
```

```python
X_test_scaled = pd.DataFrame(X_test_scaled,columns=X_test.columns)
X_test_scaled
```

## Building the Models

## Logistic Regression Model

We'll be trying to predict a classification- Dead or not dead. Let's begin our understanding of implementing Logistic Regression in Python for classification.

```python
from sklearn.linear_model import LogisticRegression
```

```python
#Logictic Regression Model

logmodel = LogisticRegression()
logmodel.fit(X_trained_scaled,y_train)
```

```python
logpredictions = logmodel.predict(X_test_scaled)
```

We can check precision,recall,f1-score using classification report!

```python
#Logistic Regression Output
print(classification_report(y_test,logpredictions))
```

Creating the Confusion Metric

```python
#Logistic Regression Confusion matrix Output
z=confusion_matrix(y_test,logpredictions)
fig,ax=plt.subplots(figsize=(8,8))
ax.imshow(z)
ax.grid(False)
ax.xaxis.set(ticks=(0,1),ticklabels=('Predicted 0s','Predicted 1s'))
ax.yaxis.set(ticks=(0,1),ticklabels=('Actual 0s','Actual 1s'))
ax.set_ylim(1.5, -0.5)
for i in range(2):
    for j in range(2):
        ax.text(j,i,z[i,j],ha='center',va='center',color='red',fontsize=25)
plt.show()
```

## Random Forest Classifier Model

We'll be trying to predict a classification- Dead or not dead. Let's begin our understanding of implementing Random Forest classifier in Python for classification.

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
#RandomForestClassifier Model

RFmodel = RandomForestClassifier()
RFmodel.fit(X_trained_scaled,y_train)
```

```python
RFpredictions = RFmodel.predict(X_test_scaled)
```

We can check precision,recall,f1-score using classification report!

```python
#RandomForest output
print(classification_report(y_test,RFpredictions))
```

Creating the Confusion Metric

```python
#RandomForest Confusion Matrix output
z=confusion_matrix(y_test,RFpredictions)
fig,ax=plt.subplots(figsize=(8,8))
ax.imshow(z)
ax.grid(False)
ax.xaxis.set(ticks=(0,1),ticklabels=('Predicted 0s','Predicted 1s'))
ax.yaxis.set(ticks=(0,1),ticklabels=('Actual 0s','Actual 1s'))
ax.set_ylim(1.5, -0.5)
for i in range(2):
    for j in range(2):
        ax.text(j,i,z[i,j],ha='center',va='center',color='red',fontsize=25)
plt.show()
```

## KNeighbors Classifier Model

We'll be trying to predict a classification- Dead or not dead. Let's begin our understanding of implementing KNeighbours Classifier in Python for classification.

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#KNeighborsClassifier Model

from sklearn import neighbors
KNNmodel = KNeighborsClassifier(n_neighbors=3)
KNNmodel.fit(X_trained_scaled,y_train)
```

```python
KNNpredictions = KNNmodel.predict(X_test_scaled);
```

We can check precision,recall,f1-score using classification report!

```python
#KNeighbour Output
print(classification_report(y_test,KNNpredictions))
```

Creating the Confusion Metric

```python
#KNeighbour Confusion matrix Output
z=confusion_matrix(y_test,KNNpredictions)
fig,ax=plt.subplots(figsize=(8,8))
ax.imshow(z)
ax.grid(False)
ax.xaxis.set(ticks=(0,1),ticklabels=('Predicted 0s','Predicted 1s'))
ax.yaxis.set(ticks=(0,1),ticklabels=('Actual 0s','Actual 1s'))
ax.set_ylim(1.5, -0.5)
for i in range(2):
    for j in range(2):
        ax.text(j,i,z[i,j],ha='center',va='center',color='red',fontsize=25)
plt.show()
```
